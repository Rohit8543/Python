### relplot function

It provides access to several different axes-level functions that show the relationship between two variables with semantic mappings of subsets. The kind parameter selects the underlying axes-level function to use:

* scatterplot() (with kind=”scatter”)
* lineplot() (with kind=”line”)

```python
sns.relplot(data=df, x="total_bill", y="tip",kind="line")
```

```python
sns.relplot(data=df, x="total_bill", y="tip", hue="day")
```

```python
sns.relplot(data=df, x="total_bill", y="tip", hue="sex", col="day", col_wrap=2)
```

```python
sns.relplot(data=df, x="size", y="tip",kind="line",ci=None)
```

#### Bar Plot
Seaborn supports many types of bar plots

```python
# Vertical barplot

sns.barplot(x = 'size', y = 'tip', data = df,
            palette = 'PuRd',ci=None 
            )
```

```python
# Horizontal barplot
# To draw a horizontal plot pass ‘h’ to the parameter, orient of the barplot function

sns.barplot(x = 'tip', y = 'day', data = df,
            palette = 'PuRd', orient = 'h', ci=None
            );
```

```python
# Count plot
#The count plot can be thought of as a histogram across a categorical variable.

sns.countplot(x = 'size', hue = 'day', data = df, palette = 'magma');
```

```python
sns.countplot(x = 'size', hue = 'smoker', data = df, palette = 'PiYG');
```

#### Point Plot

Point plot is used to show point estimates and confidence intervals using scatter plot glyphs. A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.

Point plots can be more useful than bar plots for focusing comparisons between different levels of one or more categorical variables.

```python
sns.pointplot(x="day", y="tip", data=df)
```

```python
sns.pointplot(x="time", y="total_bill", hue="smoker",
                   data=df, palette="Accent")
```

#### Joint Plot

```python
sns.set_style("dark")
sns.jointplot(x='total_bill', y='tip',data=df)
```

```python
# Add regression line to scatter plot and kernel density estimate to histogram
sns.jointplot(x='total_bill', y='tip', data=df, kind='reg') 
```

```python
# Display kernel density estimate instead of scatter plot and histogram
sns.jointplot(x='total_bill', y='tip', data=df, kind='kde')
```

```python
# Display hexagons instead of points in scatter plot
sns.jointplot(x='total_bill', y='tip', data=df, kind='hex') 
```

#### pairplot

pairplot will plot pairwise relationships across an entire dataframe (for the numerical columns) and supports a color hue argument (for categorical columns). 

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

### stripplot and swarmplot

```python
sns.stripplot(x="day", y="total_bill", data=df)
```

```python
#  jitter to reduce overplotting

sns.stripplot(x="day", y="total_bill", data=df,jitter=False)
```

```python
sns.stripplot(x="day", y="total_bill", data=df, jitter=False, 
              hue='sex', palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=df,jitter=True,
              hue='sex',palette='Set1',dodge=True)
```

```python
sns.swarmplot(x="day", y="total_bill", data=df)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',
              data=df, palette="Set1", dodge=True)
```
